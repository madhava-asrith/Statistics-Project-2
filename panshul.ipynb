```python

import numpy as np
from scipy.stats import t

# Load datasets
indusIndBk = pd.read_csv("./archive/INDUSINDBK.csv")
iciciBk = pd.read_csv("./archive/ICICIBANK.csv")  # Replace with another NIFTY company

# Extract 'Close' prices for both populations
close_indusind = indusIndBk['Close'].dropna()
close_icici = iciciBk['Close'].dropna()


# Population mean 
mean_indusind, mean_icici = np.mean(close_indusind), np.mean(close_icici)
var_indusind, var_icici = np.var(close_indusind, ddof=1), np.var(close_icici, ddof=1)


# Take two samples of size 100 from close_icici and close_indusind
sample_icici = close_icici.sample(n=100, random_state=42)
sample_indusind = close_indusind.sample(n=100, random_state=42)



# Compute descriptive statistics
mean_indusind, mean_icici = np.mean(close_indusind), np.mean(close_icici)
var_indusind, var_icici = np.var(close_indusind, ddof=1), np.var(close_icici, ddof=1)
n_indusind, n_icici = len(close_indusind), len(close_icici)

# Compute degrees of freedom (Welch-Satterthwaite approximation)
numerator = (var_indusind/n_indusind + var_icici/n_icici)**2
denominator = ((var_indusind/n_indusind)**2 / (n_indusind - 1)) + ((var_icici/n_icici)**2 / (n_icici - 1))
degrees_of_freedom = numerator / denominator

# Compute standard error and critical value
std_error = np.sqrt(var_indusind/n_indusind + var_icici/n_icici)
t_critical = t.ppf(0.975, degrees_of_freedom)  # Two-tailed test at 95% confidence level

# Compute confidence interval
mean_diff = mean_indusind - mean_icici
ci_lower = mean_diff - t_critical * std_error
ci_upper = mean_diff + t_critical * std_error

print(f"Mean Difference: {mean_diff:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

```python
import matplotlib.pyplot as plt

# Plot mean closing prices for both companies
plt.bar(['IndusIndia Bank', 'ICICI Bank'], [mean_indusind, mean_icici], color=['blue', 'green'])
plt.ylabel('Mean Closing Price')
plt.title('Comparison of Mean Closing Prices')
plt.show()
```
